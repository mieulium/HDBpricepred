#### adding closest school distance to dataframe

```python
def degreesToRadians(degrees):
    """
    returns degrees in radians
    
    Parameters
    ----------
    degrees: input degrees
    
    Returns
    -------
    radian: converted from degrees
    """
    return degrees * math.pi / 180

def distanceInKmBetweenEarthCoordinates(point_a, point_b):
    """
    returns distance in km between two geo-coordinates
    
    Parameters
    ----------
    point_a: latitude and longitude of 1st point in 'lat, long' format
    point_b: latitude and longitude of 2nd point in 'lat, long' format
    
    Returns
    -------
    distance in km between the two latitude and longitude
    """
    if type(point_a) == float or type(point_b) == float:
        return np.nan
    
    else:
        lat_a = float(point_a.split(",")[0])
        long_a = float(point_a.split(",")[1])
        
        lat_b = float(point_b.split(",")[0])
        long_b = float(point_b.split(",")[1])
        
        earthRadiusKm = 6371
        
        dLat = degreesToRadians(lat_b-lat_a)
        dLon = degreesToRadians(long_b-long_a)
        
        lata = degreesToRadians(lat_a)
        latb = degreesToRadians(lat_b)

        a = math.sin(dLat/2) * math.sin(dLat/2) + \
        math.sin(dLon/2) * math.sin(dLon/2) * \
        math.cos(lata) * math.cos(latb)\
        
        c = 2 * math.atan2(math.sqrt(a), math.sqrt(1-a))
        
        return earthRadiusKm * c
```

```python
closest_psch_dist = []

#finding the closest school distance
for x in allprice['latlong']:
    lowest = 9999
    if x == np.nan: 
        lowest = np.nan
        closest_psch_dist.append(lowest)
    else:
        for s in school.loc[school['mainlevel_code'] == 'PRIMARY']['latlong']:
            dist = distanceInKmBetweenEarthCoordinates(s,x)
            if dist >= lowest: 
                next

            else:
                lowest = dist

        closest_psch_dist.append(lowest)
    
```

```python
## defining a function to perform ANOVA through the list 
def ANOVA_list(searchlist, dataframe, resultdict):
    a = {}
    for var in searchlist:
        var_unique_list = []
        count = 0
        resultdict[var] = {}
        for i in range(len(dataframe[var].unique())):
            count += 1 
            if pd.isnull(dataframe[var].unique()[i]) == True:
                next
            else:
                var_unique_list.append(dataframe[var].unique()[i])
                dataframe[var].unique()[i] = dataframe[dataframe[var]==\
                        (dataframe[var].unique()[i])]['resale_price']
        for cat in var_unique_list:
            if pd.isnull(cat) == True:
                next
            else:
                resultdict[var][str(cat)] = \
                dataframe['resale_price'][dataframe[var]\
                                       == cat].mean()
        f, p = stats.f_oneway(*(dataframe['resale_price'][dataframe[var]\
                                    == cat] for cat in var_unique_list))
        resultdict[var]['statistic'] = f
        resultdict[var]['P'] = p 
    return resultdict
```
